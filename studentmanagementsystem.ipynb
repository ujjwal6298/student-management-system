```python
from tkinter import *
from tkinter import messagebox,ttk
import sqlite3

class Magement:
    def __init__(self,root):
        self.root=root
        self.condition=True
        
        self.entry_frame=LabelFrame(self.root,text='Enter Data ',fg='gold',bg='gray',font=('arial',10,'bold'),bd=5,relief=RIDGE)
        self.entry_frame.place(x=0,y=0,width=300,height=235)
        
        Label(self.entry_frame,text='Name :',width=11,font=('arial',10,'bold'),bd=4,relief=GROOVE).grid(row=0,column=0,ipadx=10,ipady=8)
        Label(self.entry_frame,text='Roll no :',width=11,font=('arial',10,'bold'),bd=4,relief=GROOVE).grid(row=1,column=0,ipadx=10,ipady=8)
        Label(self.entry_frame,text='Course :',width=11,font=('arial',10,'bold'),bd=4,relief=GROOVE).grid(row=2,column=0,ipadx=10,ipady=8)
        Label(self.entry_frame,text='Age :',width=11,font=('arial',10,'bold'),bd=4,relief=GROOVE).grid(row=3,column=0,ipadx=10,ipady=8)
        Label(self.entry_frame,text='DOB :',width=11,font=('arial',10,'bold'),bd=4,relief=GROOVE).grid(row=4,column=0,ipadx=10,ipady=8)
        
        
        #enter boxes...
        self.name_ent=Entry(self.entry_frame,width=20,font=('arial',10),bd=3)
        self.name_ent.grid(row=0,column=1,ipadx=10,ipady=8)
        
        self.roll_ent=Entry(self.entry_frame,width=20,font=('arial',10),bd=3)
        self.roll_ent.grid(row=1,column=1,ipadx=10,ipady=8)
        
        self.course_ent=Entry(self.entry_frame,width=20,font=('arial',10),bd=3)
        self.course_ent.grid(row=2,column=1,ipadx=10,ipady=8)
        
        self.age_ent=Entry(self.entry_frame,width=20,font=('arial',10),bd=3)
        self.age_ent.grid(row=3,column=1,ipadx=10,ipady=8)
        
        self.dob_ent=Entry(self.entry_frame,width=20,font=('arial',10),bd=3)
        self.dob_ent.grid(row=4,column=1,ipadx=10,ipady=8)
        
        #button frame....
        btn_frame=LabelFrame(self.root,text='Controls ',fg='gold',bg='gray',font=('arial',10,'bold'),bd=5,relief=RIDGE)
        btn_frame.place(x=300,y=0,width=130,height=235)
        
        
        Button(btn_frame,text='Save',font=('arial',10,'bold'),bd=3,width=8,command=self.save_data).grid(row=0,column=0,ipadx=10,padx=12,pady=4)
        
        Button(btn_frame,text='Update',font=('arial',10,'bold'),bd=3,width=8,command=self.update_data).grid(row=1,column=0,ipadx=10,padx=12,pady=4)
       
        Button(btn_frame,text='Delete',font=('arial',10,'bold'),bd=3,width=8,command=self.delete_data).grid(row=2,column=0,ipadx=10,padx=12,pady=4)
        
        Button(btn_frame,text='Clear',font=('arial',10,'bold'),bd=3,width=8,command=self.clear_data).grid(row=3,column=0,ipadx=10,padx=12,pady=4)
        
        
        self.search_ent=Entry(btn_frame,width=15,bd=3)
        self.search_ent.grid(row=4,column=0,ipadx=5)
        
        self.search_btn=Button(btn_frame,text='Search',font=('arial',10,'bold'),bd=3,width=8,command=self.search_data)
        self.search_btn.grid(row=5,column=0,ipadx=10,padx=12,pady=4)
        
        #treeview and scrollbar...
        scroll_y=Scrollbar(self.root,orient=VERTICAL,bg='gray',bd=3)
        col=('name','roll','course','age','dob')        
        self.student_table=ttk.Treeview(self.root,column=col,yscrollcommand=scroll_y.set)
        scroll_y.place(x=410,y=235,height=150)
        scroll_y.config(command=self.student_table.yview)
        
        self.student_table.heading('name',text='Name')
        self.student_table.heading('roll',text='Roll')
        self.student_table.heading('course',text='Course')
        self.student_table.heading('age',text='Age')
        self.student_table.heading('dob',text='DOB')
        
        self.student_table['show']='headings'
        self.student_table.column('name',width=20,anchor='center')
        self.student_table.column('roll',width=5,anchor='center')
        self.student_table.column('course',width=10,anchor='center')
        self.student_table.column('age',width=5,anchor='center')
        self.student_table.column('dob',width=10,anchor='center')
        
        self.student_table.bind('<ButtonRelease-1>',self.get_cursor)
        
        
        self.student_table.place(x=0,y=235,width=410,height=150)
        self.fetch_data()
    def clear_data(self):
        self.name_ent.delete(0,END)
        self.roll_ent.delete(0,END)
        self.course_ent.delete(0,END)
        self.age_ent.delete(0,END)
        self.dob_ent.delete(0,END)
        self.search_ent.delete(0,END)
        
    def get_cursor(self,event):
        cursor_row=self.student_table.focus()
        contents=self.student_table.item(cursor_row)
        row=contents['values']
        
        self.name_ent.insert(0,row[0])
        self.roll_ent.insert(0,row[1])
        self.course_ent.insert(0,row[2])
        self.age_ent.insert(0,row[3])
        self.dob_ent.insert(0,row[4])
        self.search_ent.insert(0,row[1])
    def save_data(self):
        try:
            
            conn=sqlite3.connect('student_database.db')
            sqlite_save_query="""INSERT INTO record (name,roll,course,dob) VALUES(?,?,?,?,?) """
            data_tuple=(self.name_ent.get(),self.roll_ent.get(),self.course_ent.get(),self.age_ent.get())
            
            cur=conn.cursor()
            cur.execute(sqlite_save_query,data_tuple)
            conn.commit()
            conn.close()
            
            self.clear_data()
            self.fetch_data()
            messagebox.showinfo('Python says','Data is inserted')
        except Exception as e:
            messagebox.showerror('Python says',e)
    
    def fetch_data(self):
        try:
            conn=sqlite3.connect('student_database.db')
            sqlite_select_query="""SELECT * from record """
            cur=conn.cursor()
            cur.execute(sqlite_select_query)
            
            data=cur.fetchall()
            if len(data) !=0:
                self.student_table.delete(*self.student_table.get_children())
                for row in data:
                    self.student_table.insert('',END,value=row)
            conn.close()
        except Exception as e:
            messagebox.showerror('Python says',e)
    
    def update_data(self):
        
        try:
            
            conn=sqlite3.connect('student_database.db')
            sqlite_update_query="""UPDATE record set name=?,roll=?,course=?,age=>,dob=? where roll =?"""+self.search_ent.get()
            data_tuple= data_tuple=(self.name_ent.get(),self.roll_ent.get(),self.course_ent.get(),self.age_ent.get())
            cur.execute(sqlite_update_query,data_tuple)
            conn.commit()
            conn.close()
            messagebox.showinfo('python says','Data is Updated')
            self.clear_data()
            self.fetch_data()
        except Exception as e:
            messagebox.showerror('Python says',e)
    
    
    def delete_data(self):
        try:
            
            conn=sqlite3.connect('student_database.db')
            sqlite_delete_query="""DELETE from record where roll"""+self.search_ent.get()
            conn.execute(sqlite_delete_query)
            conn.commit()
            conn.close()
            messagebox.showinfo('python says','Data is Deleted')
            self.clear_data()
            self.fetch_data()
        except Exception as e:
            messagebox.showerror('Python says',e)
    
    
    def search_data(self):
        if(self.condition):
            
            try:
                
                conn=sqlite3.connect('student_database.db')
                sqlite_search_query="""SELECT * from record  where roll ="""+self.search_ent.get()
                cur=conn.cursor()
                cur.execute(sqlite_search_query)
            
                data=cur.fetchone()
                if len(data) !=0:
                    self.student_table.delete(*self.student_table.get_children())
                    self.student_table.insert('',END,value=data)
                    self.condition=False
                    self.search_btn['text']='show All'
                conn.close()
            except Exception as e:
                messagebox.showerror('Python says',e)
                
        else:
            self.condition=True
            self.search_btn['text']='Search'
            self.fetch_data() 
            
            
        
        
        
root=Tk()
root.title('Student Management')
root.geometry('430x385')
root.resizable(0,0)
root.configure(bg='gray')
Magement(root)
root.mainloop()
```
